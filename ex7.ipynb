```python
#Exercise 11.1
# Counting words in blakepoems.txt.
fp = open( "blakepoems.txt",'r')
d={}
text=fp.read()
c=text.lower()
lst=c.split()
for i in lst:
    d[i]=lst.count(i)


    
print(d)
```

```python
#Exercise 11.2
fp = open( "blakepoems.txt",'r')
d={}
text=fp.readline()
count=0
while text:
    content=text.lower()
    lst=content.split()
    for i in lst:
        if i not in d:
            d[i]=0
        d[i]+=1

    text=fp.readline()
print(d)  
```
